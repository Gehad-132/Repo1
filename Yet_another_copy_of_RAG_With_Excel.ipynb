```python
# Build a Function For Excel data
def prepare_excel(file_path):
  loader = UnstructuredExcelLoader(file_path)
  data = loader.load()

  # Split the text into chunks
  text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 300,
    chunk_overlap = 50
  )
  chunks = text_splitter.split_documents(data)

  # Create embeddings with the OpenAI(HuggingFaceEmedding) model
  embedding_model =  HuggingFaceEmbeddings(
    model_name= "all-MiniLM-L6-v2"
  )
  db = FAISS.from_documents(chunks,embedding_model)

  return db
```

```python
# Build a function for the RAG
def rag(db, query, k = 2):
  #Try the retrieval system
  output_retrieval = db.similarity_search(query, k = k)

  # Merge the context
  output_retrieval_merged = "\n".join([doc.page_content for doc in output_retrieval])

  # Define the prompt
  prompt = f"""
  Based on this context: {output_retrieval_merged}
  Answer the following question: {query}
  If you don't have information on the answer, say you don't know.
  """

  # Call the openAI API with Langchain
  pipe = pipeline(
    "text2text-generation",
    model= "google/flan-t5-large",
    max_new_tokens= 256
    )
  llm = HuggingFacePipeline(pipeline=pipe)
  response_text = llm.invoke(prompt)

  return display(Markdown(response_text))

```

```python
# Test
db = prepare_excel("Reviews.xlsx")
```

```python
rag(db, "Prepare the top 3 improvement feedbacks and justify with data from the reviews")
```

```python
# Import libraries
from langchain_community.document_loaders import UnstructuredPDFLoader
from pdfminer import psparser
```

```python
# Build Function for PDF
def prepare_pdf(file_path):
  loader = UnstructuredPDFLoader(file_path)
  data = loader.load()

  # Split the text into chunks
  text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 300,
    chunk_overlap = 50
  )
  chunks = text_splitter.split_documents(data)

  # Create embeddings with the OpenAI(HuggingFaceEmedding) model
  embedding_model =  HuggingFaceEmbeddings(
    model_name= "all-MiniLM-L6-v2"
  )
  db = FAISS.from_documents(chunks,embedding_model)

  return db
```
